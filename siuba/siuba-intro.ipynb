# Siub - Intro

```python
from siuba import _, group_by, summarize
from siuba.data import mtcars
```

```python
(mtcars
  >> group_by(_.cyl)
  >> summarize(avg_hp = _.hp.mean())
  )
```

## Filter Rows

```python
from siuba import _, filter, group_by
from siuba.data import mtcars

mtcars >> filter(_.cyl == 4, _.gear == 5)
```

```python
mtcars >> filter((_.cyl == 4) | (_.gear == 5))
```
